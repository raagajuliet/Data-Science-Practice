```python
# Missing values Nan - if there are lot of rows with missing values for a particular column we can drop the column, or if a record 
# is missing many values we can drop the record. Other ways are to average the rest of the responses for that column.

missing = np.nan
s1 = pd.Series(['r1', 'r2', missing, 'r3', 'r4', 'r5', 'r6', missing, 'r8'])
s1
```

```python
#to find missing values
s1.isnull()
```

```python
df.isnull()
```

```python
df.iloc[3:6, 0] = missing
df.iloc[1:5, 5] = missing
df
```

```python
# to fill all the Nan with a specific value - fillna
df_filled = df.fillna(0)
df_filled
```

```python
#to fill all the Nan in a particular column with a particular value use dictionary
df_filled = df.fillna({'c1': 0.1, 'c6':1.25})
df_filled
```

```python
#we can use a fill forward method to fill Nan with the last number in the column
df_filled = df.fillna(method='ffill')
df_filled
```

```python
#groupby
cars = pd.read_csv('mtcars.csv')
```

```python
cars.head()
```

```python
cars.columns=['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
gr = cars.groupby(cars['cyl'])
gr.mean()
```

```python
am = cars.groupby(cars['am']).mean()
am['mpg'].sort_values(ascending = False)
```

```python
subset = cars[['mpg', 'disp', 'hp', 'wt']]
subset
```

```python
gr1 = cars.groupby(cars['cyl'])
gr1.sum()
```
